```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('/content/archive.zip')
```

```python
credit_card_data.head()
```

each v means transaction

class stands for crct or bad transaction

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

this data is highly unbalanced

0-normal
1-fraud

```python
crct=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(crct.shape)
print(fraud.shape)
```

```python
crct.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
crct_sample=crct.sample(n=492)
```

```python
new_data=pd.concat([crct_sample,fraud],axis=0)
```

```python
#axis=0 means row wise
#axis 1 means column wise
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
#
```

```python
x=new_data.drop(columns='Class',axis=1)
y=new_data['Class']
```

```python
y.head()
```

```python
x.head()
```

```python
x.info()
```

```python
print(x)
```

```python
#split data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

stratify means it will distipute data evenly

```python
print(x_train)
```

```python
print(x_test.shape)
```

```python
print(x_train.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
```

```python
traing_data_accuracy=accuracy_score(x_train_pred,y_train)
```

```python
print(traing_data_accuracy)
```

```python
x_test_pred=model.predict(x_test)
```

```python
x_test_accuracy=accuracy_score(x_test_pred,y_test)
```

```python
print(x_test_accuracy)
```
